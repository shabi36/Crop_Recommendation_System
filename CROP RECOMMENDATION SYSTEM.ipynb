```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder , StandardScaler
from sklearn.metrics import  accuracy_score , confusion_matrix , precision_score
```

```python

from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier,ExtraTreesClassifier
from sklearn.svm import SVC
```

```python
dataset.head(1)
```

```python

knn= KNeighborsClassifier (n_neighbors=8 )

pipe = Pipeline ([
    
     ("step1" , StandardScaler()),
    ("step2" , knn)
])


pipe.fit (x_train , y_train)

y_predict = pipe.predict (x_test)


print("\nTRAIN ",pipe.score(x_train,y_train))
print("TEST", pipe.score(x_test,y_test))



print("\nACCURACY SCORE", accuracy_score(y_test , y_predict))
print("PRECISION SCORE", precision_score(y_test , y_predict , average=None) )
```

```python
dt= DecisionTreeClassifier (max_depth= 10)

pipe = Pipeline ([
  
     ("step1" , StandardScaler()),
    ("step2" , dt)
])


pipe.fit (x_train , y_train)

y_predict = pipe.predict (x_test)


print("\nTRAIN ",pipe.score(x_train,y_train))
print("TEST", pipe.score(x_test,y_test))



print("\nACCURACY SCORE", accuracy_score(y_test , y_predict))
print("PRECISION SCORE", precision_score(y_test , y_predict , average=None) )
```

```python
rf= RandomForestClassifier (n_estimators=300,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=25)

pipe = Pipeline ([
    
     ("step2" , StandardScaler()),
    ("step3" , rf)
])


pipe.fit (x_train , y_train)

y_predict = pipe.predict (x_test)



print("\nTRAIN ",pipe.score(x_train,y_train))
print("TEST", pipe.score(x_test,y_test))



print("\nACCURACY SCORE", accuracy_score(y_test , y_predict))
print("PRECISION SCORE", precision_score(y_test , y_predict , average=None) )
```

```python
et=  ExtraTreesClassifier (n_estimators=500,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=15,
                             bootstrap=True)

pipe = Pipeline ([
   
     ("step2" , StandardScaler()),
    ("step3" , et)
])


pipe.fit (x_train , y_train)

y_predict = pipe.predict (x_test)



print("\nTRAIN ",pipe.score(x_train,y_train))
print("TEST", pipe.score(x_test,y_test))



print("\nACCURACY SCORE", accuracy_score(y_test , y_predict))
print("PRECISION SCORE", precision_score(y_test , y_predict , average=None) )
```

```python
from sklearn.ensemble import VotingClassifier



dt= DecisionTreeClassifier (max_depth= 10)


rf= RandomForestClassifier (n_estimators=300,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=25)

gd=  GradientBoostingClassifier (n_estimators=200)


et=  ExtraTreesClassifier (n_estimators=500,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=15,
                             bootstrap=True)


knn= KNeighborsClassifier (n_neighbors=8 )



vc = VotingClassifier ([ ("rf" , rf ), ("gd" , gd) , ("et" , et) , ("dt" , dt) , ("knn" , knn)] , weights = [5,5,1,1,1])


pipe_vc = Pipeline ([
    
      ("step1" , StandardScaler()),
    ("step2" , vc)
])

```

```python
y_predict = pipe_vc.predict (x_test)


print("\nTRAIN ",pipe.score(x_train,y_train))
print("TEST", pipe.score(x_test,y_test))



print("\nACCURACY SCORE", accuracy_score(y_test , y_predict))
print("PRECISION SCORE", precision_score(y_test , y_predict , average=None) )
```
